```python
import numpy as np

x = np.array([0.0, 10.0, 100.0]) / 100.0
y = np.array([32.0, 50.0, 212.0])
w = 3.0
b = 3.0

for _ in range(30):
    yh = x * w + b
    l = (yh - y)**2
    dldyh = 2*(yh - y)
    dyhdw = x
    dyhdb = 1
    dldw = dldyh * dyhdw
    dldb = dyhdb * dldyh

    lr = 0.2
    w -= lr * dldw.sum()
    b -= lr * dldb.sum()

    print(l.sum())
```

```python
x = [0,10,100]
y = [32,50,212]
w = 3
b = 3


for i in range(100000):
    yh0 = x[0]*w+b
    yh1 = x[1]*w+b
    yh2 = x[2]*w+b

    # pre loss
    l0 = (yh0-y[0])**2
    l1 = (yh1-y[1])**2
    l2 = (yh2-y[2])**2

    # Loss
    mse = (l0 + l1 + l2) / len(x)

    # Grad
    dmsedmse = 1

    dmsedl2 = 1/len(x)
    dmsedl1 = 1/len(x)
    dmsedl0 = 1/len(x)

    dl2dyh2 = 2*(yh2-y[2])
    dl1dyh1 = 2*(yh1-y[1])
    dl0dyh0 = 2*(yh0-y[0])

    dyh2dw = x[2]
    dyh1dw = x[1]
    dyh0dw = x[0]

    dyh2db = 1
    dyh1db = 1
    dyh0db = 1

    dmsedw = (dmsedmse * dmsedl2 * dl2dyh2 * dyh2dw
            + dmsedmse * dmsedl1 * dl1dyh1 * dyh1dw
            + dmsedmse * dmsedl0 * dl0dyh0 * dyh0dw)

    dmsedb = (dmsedmse * dmsedl2 * dl2dyh2 * dyh2db
            + dmsedmse * dmsedl1 * dl1dyh1 * dyh1db
            + dmsedmse * dmsedl0 * dl0dyh0 * dyh0db)

    # Update
    lr = 0.0002
    w = w - lr*dmsedw
    b = b - lr*dmsedb
    
    print(mse)
```
