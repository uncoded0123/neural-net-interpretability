```python
# batch size 3, full batch not mini
x = [0,10,100]
y = [32,50,212]
w = 3
b = 3


for i in range(30000):
    yh0 = x[0]*w+b # single neuron, first prediction
    yh1 = x[1]*w+b
    yh2 = x[2]*w+b

    # pre loss
    l0 = (yh0-y[0])**2
    l1 = (yh1-y[1])**2
    l2 = (yh2-y[2])**2

    # Loss
    mse = (l0 + l1 + l2) / len(x)

    # Grad
    dmsedmse = 1 # unnecessary but intuitive

    dmsedl2 = 1/len(x)
    dmsedl1 = 1/len(x)
    dmsedl0 = 1/len(x)

    dl2dyh2 = 2*(yh2-y[2])
    dl1dyh1 = 2*(yh1-y[1])
    dl0dyh0 = 2*(yh0-y[0])

    dyh2dw = x[2]
    dyh1dw = x[1]
    dyh0dw = x[0]

    dyh2db = 1
    dyh1db = 1
    dyh0db = 1

    dmsedw = (dmsedmse * dmsedl2 * dl2dyh2 * dyh2dw
            + dmsedmse * dmsedl1 * dl1dyh1 * dyh1dw
            + dmsedmse * dmsedl0 * dl0dyh0 * dyh0dw)

    dmsedb = (dmsedmse * dmsedl2 * dl2dyh2 * dyh2db
            + dmsedmse * dmsedl1 * dl1dyh1 * dyh1db
            + dmsedmse * dmsedl0 * dl0dyh0 * dyh0db)

    # Update
    lr = 0.0002
    w = w - lr*dmsedw
    b = b - lr*dmsedb
    if i%1000 == 0:
        print(mse)
```
